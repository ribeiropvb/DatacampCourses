# Sorting and subsetting

## Exercises 1 - Sorting rows

```python
import pandas as pd

homelessness = pd.read_csv(r'C:\Users\Pedro\Desktop\Pedro\Datacamp\DatacampCourses\Data Manipulation with pandas\data\homelessness.csv')
```

```python
# Sort homelessness by individuals
homelessness_ind = homelessness.sort_values('individuals')

# Print the top few rows
homelessness_ind.head()
```

```python
# Sort homelessness by descending family members
homelessness_fam = homelessness.sort_values('family_members', ascending=False)

homelessness_fam.head()
```

```python
# Sort homelessness by region, then descending family members
homelessness_reg_fam = sorted_df_mixed = homelessness.sort_values(by=['region', 'family_members'], ascending=[True, False])


# Print the top few rows
homelessness_reg_fam.head()
```

## Exercises 2 - Subsetting columns

```python
# Select the individuals column
individuals = homelessness['individuals']

individuals.head()
```

```python
# Select the state and family_members columns
state_fam = homelessness[['state','family_members']]

state_fam.head()
```

```python
# Select only the individuals and state columns, in that order
ind_state = homelessness[['individuals','state']]

ind_state.head()
```

## Exercises 3 - Subsetting rows

```python
# Filter for rows where individuals is greater than 10000
ind_gt_10k = homelessness[homelessness['individuals'] > 10000]

# See the result
ind_gt_10k
```

```python
# Filter for rows where region is Mountain
mountain_reg = homelessness[ homelessness['region'] == 'Mountain' ]

# See the result
mountain_reg
```

```python
# Filter for rows where family_members is less than 1000 
# and region is Pacific
fam_lt_1k_pac = homelessness[(homelessness['family_members'] < 1000) & (homelessness['region'] == 'Pacific')]

# See the result
fam_lt_1k_pac
```

## Exercises 4 - Subsetting rows by categorical variables

```python
# The Mojave Desert states
canu = ["California", "Arizona", "Nevada", "Utah"]

# Filter for rows in the Mojave Desert states
mojave_homelessness = homelessness[homelessness["state"].isin(canu)]

# See the result
mojave_homelessness
```
